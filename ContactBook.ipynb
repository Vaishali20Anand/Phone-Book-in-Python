```python
import sys
```

```python
#first function
def default_phonebook():
    row, col = int(input("Please enter no. of contacts: ")), 5
    
    book = []
    print(book)
    for i in range(row):
        print("\nEnter contact %d details in the following order (ONLY):" % (i+1))
        print("----------------------------------------------------------------")
        temp=[]
        for j in range(col):
            if j == 0:
                temp.append(str(input("Enter Name:* ")))
                
                #Checking blank input
                if temp[j] == "" or temp[j] == " ":
                    sys.exit("Name is mandatory. Exiting system...")
                    
            if j == 1:
                temp.append(int(input("Enter number:* ")))
                #No need to check blank becoz it is done by int function
                
            if j == 2:
                temp.append(str(input("Enter e-mail: "))) 
                if temp[j] == "" or temp[j] == " ":
                    temp[j] = None
                    
            if j == 3:
                temp.append(str(input("Enter DOB(dd/mm/yyyy): "))) 
                if temp[j] == "" or temp[j] == " ":
                    temp[j] = None
                    
            if j == 4:
                temp.append(str(input("Enter Category(Family/Friends/Work/Others): "))) 
                if temp[j] == "" or temp[j] == " ":
                    temp[j] = None
                    
        book.append(temp)
        
    print(book)
    return book
```

```python
def menu():
    print("---------------------------------------------------------------")
    print("PHONE DIRECTORY",flush = False)
    print("---------------------------------------------------------------")
    print("These are the operations you can perform")
    print("1. Add a new contact")
    print("2. Remove an existing contact") 
    print("3. Delete all contacts") 
    print("4. Search for a contact") 
    print("5. Display all contacts") 
    print("6. Exit contact book")
    
    choice = int(input("Please enter your choice from given options: "))
    return choice
```

```python
def add(pb):
    dip=[]
    for i in range(len(pb[0])):
        if i==0:
            dip.append(str(input("Enter name: ")))
        if i==1:
            dip.append(str(input("Enter number: ")))
        if i==2:
            dip.append(str(input("Enter email: ")))
        if i==3:
            dip.append(str(input("Enter DOB: ")))
        if i==4:
            dip.append(str(input("Enter category: ")))
        
    pb.append(dip)
    return pb
```

```python
def remove(pb):
    query = str(input("Enter name of contact to be removed: "))
    temp = 0
    
    for i in range(len(pb)):
        if query == pb[i][0]:
            temp +=1
            
            print(pb.pop(i))
            print("This contact is removed")
        
            return pb
        
        if temp == 0:
            print("INVALID QUERY!!")
            return pb
```

```python
def delete(pb):
    return pb.clear()
```
